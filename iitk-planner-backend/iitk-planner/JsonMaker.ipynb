```python
import pandas as pd
import re

# Assuming your dataframe is named 'df'
# Replace the column names below with the appropriate ones from your dataframe
day_columns = ['day1', 'day2', 'day3', 'day4', 'day5', 'day7', 'day8', 'day9', 'day10', 'day11']

# Filter the dataframe to include only the relevant columns
df_relevant = df[['Course Name', 'Course Code', 'Instructor', 'Credits'] + day_columns]

# Create a list to store the JSON objects for each course
json_data = []

# Iterate through each row in the dataframe
for index, row in df_relevant.iterrows():

    row = df.iloc[index]
    # Extract relevant data from the row
    course_name = row['Course Name']
    course_code = row['Course Code']
    instructor = row['Instructor']
    credits = str(row['Credits'])
    
    # Create a list to store timings for the current course
    timings = []
    
    # Iterate through the day columns and their corresponding start and end time columns
    for day in day_columns:
        if pd.notnull(row[day]):  # Check if the day exists
            # Extract the day number from the column name using regular expression
            day_index = int(re.findall(r'\d+', day)[0])
            start_time_col = f'start_time_{day_index}'
            end_time_col = f'end_time_{day_index}'
            
            # Check if the start and end time columns exist and are not empty
            if pd.notnull(row[start_time_col]) and pd.notnull(row[end_time_col]):
                day_of_week = row[day]
                start_time = row[start_time_col]
                end_time = row[end_time_col]
                
                # Create a dictionary for each timing
                timing_dict = {
                    "id": len(timings) + 1,  # This is just an incremental ID for each timing
                    "startTime": start_time,
                    "endTime": end_time,
                    "day": day_of_week
                }
                
                # Append the timing dictionary to the 'timings' list
                timings.append(timing_dict)
    
    # Create a dictionary for the current course
    course_dict = {
        "id": index + 1,  # You can use 'index + 1' as the ID for each course (1-based indexing)
        "name": course_name,
        "courseCode": course_code,
        "professor": instructor,
        "credits": credits,
        "timings": timings
    }
    
    # Append the course dictionary to the 'json_data' list
    json_data.append(course_dict)

# 'json_data' now contains the JSON objects for each course
# You can export this list to a JSON file if needed

print(json_data)
import json

# Export the JSON data to a file (optional)
with open("output.json", "w") as f:
    json.dump(json_data, f)
```
